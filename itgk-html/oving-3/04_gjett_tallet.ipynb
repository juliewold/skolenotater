I denne oppgaven skal du lage et program som genererer et tilfeldig heltall i et gitt intervall, og deretter lar brukeren gjette på hvilket tall dette er. Dette bør gjøres ved bruk av løkker.

I denne oppgaven må du benytte deg av biblioteket random. Derfor skal koden din starte med `import random`.

Eksempel på bruk av `random.randint()`:

```python
tilfeldig_tall = random.randint(1, 50)  # Gir deg et tilfeldig tall fra og med 1, til og med 50
tall = random.randint(a, b)  # Gir deg et tilfeldig tall mellom a og b. (Fra og med a, til og med b)
```

## a)

Be brukeren om å velge en nedre og en øvre grense for tall han eller hun skal gjette på. Lagre disse to opplysningene i to variabler.

***Skriv koden din her:***

````python
# Min kode (opg a,b,c)
import random 
# Bruker bestemmer intervallet tallet ligger i
a = int(input("Skriv en nedre grense for det tilfeldige tallet"))
b = int(input("Skriv en øvre grense for det tilfeldige tallet"))
tilfeldig_tall = random.randint(a,b) # Genereres et tilfeldig tall innenfor brukerens gitte intervall

while True: # Løkken kjører helt til bruker har gjettet riktig tall
    c = int(input("Make a guess ")) # Ber brukeren gjette et tall
    if c < tilfeldig_tall: # Hvis bruker gjetter et tall mindre enn:
        print("The correct number is higher")
    elif c > tilfeldig_tall: # Hvis bruker gjetter et tall større enn:
        print("The correct number is lower")
    else: # Om brukeren gjetter riktig tall:
        print("You guessed correct!")
        break
````
